```python
licensing['개업년'] = licensing['개업시기'].str[:4]
licensing['개업월'] = licensing['개업시기'].str[4:6]
licensing['개업일'] = licensing['개업시기'].str[6:8]
licensing
```

```python
licensing['폐업년'] = licensing['폐업시기'].str[:4]
licensing['폐업월'] = licensing['폐업시기'].str[4:6]
licensing['폐업일'] = licensing['폐업시기'].str[6:8]
licensing
```

```python
licensing.loc[licensing['개업월'] == '', '개업월'] = '00'
licensing.loc[licensing['개업일'] == '', '개업일'] = '00'
```

```python
licensing['개업년'] = licensing['개업년'].astype(int)
licensing['개업월'] = licensing['개업월'].astype(int)
licensing['개업일'] = licensing['개업일'].astype(int)
licensing['폐업년'] = licensing['폐업년'].astype(float)
licensing['폐업월'] = licensing['폐업월'].astype(float)
licensing['폐업일'] = licensing['폐업일'].astype(float)
licensing
```

```python
licensing.to_csv('local.csv', encoding='cp949')
```

```python
licensing = licensing[licensing.폐업월 < 12]
licensing = licensing[licensing.개업월 < 12]
licensing = licensing[(licensing['폐업일'] < 31) | (licensing['폐업일'] != np.nan)]
licensing = licensing[(licensing['폐업일'] < 31) | (licensing['개업일'] != np.nan)]
licensing = licensing[(licensing['폐업월'] != 6) & (licensing['폐업일'] != 31)]
licensing = licensing[(licensing['개업월'] != 6) & (licensing['개업일'] != 31)]
# licensing = licensing[(licensing['폐업일'] != 0) | (licensing['폐업일'] != np.nan)]
# licensing = licensing[(licensing['개업일'] != 0) | (licensing['개업일'] != np.nan)]
licensing = licensing[licensing.개업년 > 2000]
licensing = licensing[licensing.폐업년 > 2000]
licensing
```

```python
licensing.isnull().sum()
```

```python
licensing['폐업일자'] = licensing['폐업시기'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
licensing['개업일자'] = licensing['개업시기'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
licensing
```
