```python
import pandas as pd
import os
import openpyxl
```

```python
path = 'C:/Users/Edward/Desktop/LOCALDATA_ALL/'
xlsx = os.listdir(path)
```

```python
df = pd.DataFrame([])
for j in xlsx:
    sheet = openpyxl.load_workbook(path + j).sheetnames
    for i in sheet:
        df1 = pd.read_excel(path + j, sheet_name=i, usecols = [1, 3, 4, 5, 7, 8, 11, 12, 13, 14, 18, 19, 21, 25, 26, 27])
        df = pd.concat([df, df1])

df.to_csv("licensing.csv", index=False)
```
