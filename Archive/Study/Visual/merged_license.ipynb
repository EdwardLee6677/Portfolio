```python
import re

# 시, 군, 구 추출 함수 정의
def split_address(address):
    if not isinstance(address, str):  # 문자열이 아니면 None 반환
        return pd.Series({'시': None, '군': None, '구': None})
    
    pattern = r'(?P<si>[가-힣]+시)?\s?(?P<gun>[가-힣]+군)?\s?(?P<gu>[가-힣]+구)?'
    match = re.search(pattern, address)
    if match:
        return pd.Series({
            '시': match.group('si'),
            '군': match.group('gun'),
            '구': match.group('gu')
        })
    return pd.Series({'시': None, '군': None, '구': None})

# 데이터프레임에 적용하여 새로운 컬럼 추가
merged_df[['시', '군', '구']] = merged_df['도로명전체주소'].apply(split_address)

merged_df.head()
```

```python
# groupby 메서드를 이용하여 일자별 '개방서비스명'으로 그룹화한 뒤 '기준' 컬럼의 합계 계산
# '구'열도 추가
df_o = merged_df.groupby(['인허가일자', '개방서비스명', '구'])['기준'].sum().reset_index()
df_c = merged_df.groupby(['폐업일자', '개방서비스명', '구'])['기준'].sum().reset_index()
df_rs = merged_df.groupby(['휴업시작일자', '개방서비스명', '구'])['기준'].sum().reset_index()
df_rf = merged_df.groupby(['휴업종료일자', '개방서비스명', '구'])['기준'].sum().reset_index()
```

```python
# Grouping by '인허가일자' and '구' with all services counted separately
df_o = df_o.groupby(['인허가일자', '구', '개방서비스명']).size().unstack(level=-1, fill_value=0)
df_c = df_c.groupby(['폐업일자', '구', '개방서비스명']).size().unstack(level=-1, fill_value=0)
df_rs = df_rs.groupby(['휴업시작일자', '구', '개방서비스명']).size().unstack(level=-1, fill_value=0)
df_rf = df_rf.groupby(['휴업종료일자', '구', '개방서비스명']).size().unstack(level=-1, fill_value=0)

df_o.reset_index(inplace=True)
df_c.reset_index(inplace=True)
df_rs.reset_index(inplace=True)
df_rf.reset_index(inplace=True)
```
