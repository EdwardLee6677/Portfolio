```python
attribute = 'Salary'
Q1 = df[attribute].quantile(0.25)
Q3 = df[attribute].quantile(0.75)
IQR = Q3 - Q1
outlier_step = 1.5 * IQR
outlier = df[(df[attribute] < Q1 - outlier_step) | (df[attribute] > Q3 + outlier_step)]
outlier # 왜 교재와 다르게 안나오는지 몰라도 이상치 없음
```

```python
upper_limit = df[attribute].mean() + 3 * df[attribute].std()
lower_limit = df[attribute].mean() - 3 * df[attribute].std()

df[(df[attribute] > upper_limit) | (df[attribute] < lower_limit)] # 왜 교재와 다르게 안나오는지 몰라도 이상치 없음
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
X,y = make_classification(n_classes=2, class_sep=2, flip_y=0, weights=[0.1, 0.9], n_features=20, n_clusters_per_class=1, n_samples=1000)
print(f"Before SMOTE: {Counter(y)}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)
print(f"After SMOTE: {Counter(y_res)}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], label="Class #0", alpha=0.5, linewidth=0.15)
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], label="Class #1", alpha=0.5, color='r', linewidth=0.15)
plt.title('Original class distribution')
plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label="Class #0", alpha=0.5, linewidth=0.15)
plt.scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label="Class #1", alpha=0.5, color='r', linewidth=0.15)
plt.title('After SMOTE class distribution')
plt.legend()
plt.tight_layout()
plt.show() # plt 시각화 결과가 다르게 나오는 이유는 69번 셀에서의 random_state 떄문? 인듯
```

```python
data = {
'Education': ['High School', 'Bachelors', 'Masters', 'PhD', 'Associates', 'Bachelors', 'Masters'],
'Income': ['Low', 'Low', 'Medium', 'High', 'Low', 'Medium', 'High']}
df = pd.DataFrame(data)
df
```

```python
education_order = {'High School': 1, 'Associates': 2, 'Bachelors': 3, 'Masters': 4, 'PhD': 5}
income_order = {'Low': 1, 'Medium': 2, 'High': 3}
df['Education_encoded'] = df['Education'].map(education_order)
df['Income_encoded'] = df['Income'].map(income_order)
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
education_categories = ['High School', 'Associates', 'Bachelors', 'Masters', 'PhD']
income_categories = ['Low', 'Medium', 'High']
encoder = OrdinalEncoder(categories=[education_categories, income_categories])
df_encoded = encoder.fit_transform(df[['Education', 'Income']])
```

```python
df['Education_encoded'] = df_encoded[:, 0]
df['Income_encoded'] = df_encoded[:, 1]
df
```
