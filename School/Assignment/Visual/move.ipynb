```python
import pandas as pd
# 데이터 불러올때 헤더가 없는 경우
mv = pd.read_csv('시군구별_이동자수_20241126172236.csv', encoding='euc-kr', header=None)
mv
```

```python
#  기존 열이름 + 첫번째 행 내용 합치기
mv.columns = mv.iloc[0] + mv.iloc[1]
# 1,2행 삭제
mv = mv.drop([0,1])
mv.head()
```

```python
#1번째 열 이름 변경
mv.rename(columns={mv.columns[0]:'시점'}, inplace=True)
mv.head()
```

```python
# 시점을 인덱스로 설정
mv.set_index('시점', inplace=True)
mv.head()
```

```python
# 열이름 중 순이동자가 포함된 열만 추출
pmv = mv.filter(like='순이동', axis=1)
pmv.head()
```

```python
# 전체 열이름에서 '순이동(명)' 텍스트부분을 제거
pmv.columns = pmv.columns.str.split('순').str[0]
pmv.head()
```

```python
# 열이름 중 총전입이 포함된 열만 추출
inmv = mv.filter(like='총전입', axis=1)
inmv.columns = inmv.columns.str.split('총').str[0]
inmv.head()
```

```python
outmv = mv.filter(like='총전출', axis=1)
outmv.columns = outmv.columns.str.split('총').str[0]
outmv.head()
```

```python
pmv.to_csv('pmv.csv', encoding='utf-8')
inmv.to_csv('inmv.csv', encoding='utf-8')
outmv.to_csv('outmv.csv', encoding='utf-8')
```
