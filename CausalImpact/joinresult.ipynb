```python
# 디렉토리 경로 설정
directory_path = 'C:/Users/Edward/Desktop/PythonWorkspace/인허가/각 사건별 상황'

# 디렉토리 내의 CSV 파일 목록 가져오기
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

# 파일 목록을 기반으로 사전 생성
file_to_variable = {os.path.join(directory_path, file): file.split('.')[0] for file in csv_files}

# 생성된 사전 확인
file_to_variable
```

```python
dfs = {}
```

```python
for file_name, variable_name in file_to_variable.items():
    df = pd.read_csv(file_name, delimiter=';', header=None)
    df.columns = ['업종', variable_name, variable_name, variable_name]
    dfs[variable_name] = df
```

```python
for variable_name, df in dfs.items():
    globals()[variable_name] = df
```

```python
# 업종을 기준으로 데이터프레임 합치기
result_df = pd.concat([df.set_index('업종') for df in dfs.values()], axis=1)
first_row = result_df.iloc[0]
sorted_result_df = result_df[1:].sort_values(by='업종', ascending=True)
result_df = pd.concat([first_row.to_frame().T, sorted_result_df])
result_df.reset_index(inplace=True)
result_df = result_df.applymap(lambda x: int(x) if pd.api.types.is_numeric_dtype(x) else x)
```

```python
import openpyxl
from openpyxl.styles import PatternFill, Font
```

```python
# 엑셀 파일 열기
workbook = openpyxl.load_workbook('C:/Users/Edward/Desktop/PythonWorkspace/인허가/결과/result.xlsx')
worksheet = workbook.active

# 필터 적용 (두 번째 행)
worksheet.auto_filter.ref = 'A2:AW2'

# B2:AW189 범위의 셀 서식을 숫자로 변경
for row in worksheet.iter_rows(min_row=2, max_row=189, min_col=2, max_col=49):  # 열 범위에 따라 min_col 및 max_col 수정
    for cell in row:
        cell.number_format = '0.00'  # 숫자 서식으로 변경, 소수점 2자리까지

# B2:AW189 범위의 숫자에 대한 배경색 설정
for row in worksheet.iter_rows(min_row=2, max_row=189, min_col=2, max_col=49):  # 열 범위에 따라 min_col 및 max_col 수정
    for cell in row:
        if cell.value is not None:
            if isinstance(cell.value, (int, float)) and cell.value < 0.05:
                cell.fill = PatternFill(start_color="FFFF66", end_color="FFFF66", fill_type="solid")

# A열에 있는 특정 문자열 값을 빨간색으로 변경
target_strings = [
    '공연장', '관광공연장업', '관광사업자', '관광숙박업', '관광식당', '관광유람선업',
    '관광유흥음식점업', '관광펜션업', '국내여행업', '국내외여행업', '농어촌민박업', '숙박업',
    '스키장', '승마장업', '시내순환관광업', '썰매장업', '외국인관광도시민박업',
    '외국인전용유흥음식점업', '요트장업', '일반야영장업', '일반유원시설업', '종합여행업',
    '종합유원시설업', '한옥체험업'
]

for row in worksheet.iter_rows(min_row=2, max_row=189, min_col=1, max_col=1):  # A열만 검사
    for cell in row:
        if cell.value is not None and cell.value in target_strings:
            cell.font = Font(color="FF0000")


# 엑셀 파일 저장
workbook.save('C:/Users/Edward/Desktop/PythonWorkspace/인허가/결과/result_design.xlsx')
```
